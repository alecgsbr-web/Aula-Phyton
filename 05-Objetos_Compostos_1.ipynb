## 01 – Quais as vantagens da utilização de objetos compostos em detrimento dos objetos simples?

Um objeto simples armazena uma informação em um lugar específico, sendo assim se for necessário usar diferentes valores, eles serão armazenados em diversos lugares diferentes, usando um objeto composto, é possível armazenar diversos valores em locais sequênciais, facilitando o acesso e a comparação dos valores.
