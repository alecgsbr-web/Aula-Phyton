# Exercício 02:

Em Python, o método com nome count() faz parte do objeto da classe string. Este objeto possuí outros atributos e métodos. Qual o código que pode ser usado para verificar se count() é um método também presente no objeto da classe integer?

R: Para verificar se count() também é um método presente no objeto de classe integer você pode utilizar "count" in dir(int) ou pode testar e verificar se é possível, definindo um objeto da classe integer e usando o método count() ou usando "count" in dir('seu objeto')

```python
numero = 1
numero.count()
```

```python
numero = 1
"count" in dir(numero)
```

```python
"count" in dir(int)
```
