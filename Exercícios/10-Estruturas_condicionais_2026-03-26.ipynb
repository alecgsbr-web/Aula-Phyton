## Instancie um objeto da classe array numpy, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3. 
Quantos e quais são os valores superiores a 10?

```python
import numpy as np
array1 = np.random.normal(loc=10, scale=3, size=5)
print(array1)
ex1 = array1[array1 > 10]
print("Quantos valores são maiores que 10:")
print(len(array1))
print("Quais são os valores:")
print(ex1)
```
